```python
import re

str = "Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can."
str_word_only = re.sub('[.|,]','',str)
str_arr = str_word_only.split()

arr_oneletter = [1,5,6,7,8,9,15,16,19]
ans = {}

for i, s in enumerate(str_arr):
    if i+1 in arr_oneletter:
        ans[i+1] = s[:1]
    else:
        ans[i+1] = s[:2]

print(ans)
```

```python
str_given = "I am an NLPer"

def ngram (str):
    word_bigram = []
    letter_bigram = []

    str_split = str.split()
    letter = "".join(str_split)

    for i in range(len(letter)-1):
        letter_bigram.append(letter[i:i+2])
        
    for i in range(len(str_split)-1):
        word_bigram.append(str_split[i:i+2])

    return letter_bigram, word_bigram

print(ngram(str_given))
```

```python
x_bigram,_ = ngram("paraparaparadise")
y_bigram,_ = ngram("paragraph")

print(x_bigram)
print(y_bigram)

x_bigram = set(x_bigram)
y_bigram = set(y_bigram)


print(f"和集合： {x_bigram | y_bigram}")
print(f"積集合： {x_bigram & y_bigram}")
print(f"差集合： {x_bigram - y_bigram}")

if "se" in x_bigram:
    print("seはXに含まれる")
if "se" in y_bigram:
    print("seはYに含まれる")
```

```python
def cipher(text):
    ans_list=list()

    for c in text:
        char_add=""
        if c.islower():
            char_add=chr(219-ord(c))
        else:
            char_add=c
        ans_list.append(char_add)
    
    
    return "".join(ans_list)

output = cipher("atestz_ARddfasdfa241432")
print(output)

print(cipher(output))
```

```python
import random

def typoglycemia(text):
    t_wordlist = text.split()
    output_list = list()

    for w in t_wordlist:
        t_word = ""
        if len(w)>4:
            startc = w[0]
            endc = w[-1:]
            t_w = w[1:-1]
            t_word = startc + ''.join(random.sample(t_w,len(t_w))) +endc
        else:
            t_word = w
        
        output_list.append(t_word)
    
    return ' '.join(output_list)

print(typoglycemia("I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind ."))
```
